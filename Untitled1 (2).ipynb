```python
import pandas as pd
```

```python
data=pd.read_csv(r'C:\Users\ashmit\Downloads\USDINR_DataFile.csv')
```

```python
data.isna().sum()
```

```python
data.drop('Forieg_equit_inv_Ind_D', axis=1, inplace=True)
data.drop('Forieg_Debt_inv__Ind_D', axis=1, inplace=True)
data.drop('Tourism_IND_USD_mn_Yr', axis=1, inplace=True)
data.drop('Tourism_USA_USD_mn_Yr', axis=1, inplace=True)
data
```

```python
data.drop('Discount_rate_USA_perc_m', axis=1, inplace=True)
data.drop('Fiscal_Deficit_USA_per_Y', axis=1, inplace=True)
```

```python
data['Date']=pd.to_datetime(data['Date'])
```

```python
data['month']=data['Date'].dt.month
data['year']=data['Date'].dt.year
data['date']=data['Date'].dt.day
data.drop('Date', axis=1, inplace=True)
```

```python
for column in data.columns:
    if(column!='INR_per_USD'):
        data[column]=data[column]/data[column].max()
```

```python
#gets a random 80% of the entire set
X_train = data.sample(frac=0.9, random_state=1)
#gets the left out portion of the dataset
X_test = data.loc[~data.index.isin(X_train.index)]
data['INR_per_USD']
```

```python
y_train=X_train['INR_per_USD']
y_test=X_test['INR_per_USD']
```

```python
X_train.drop('INR_per_USD', axis=1, inplace=True)
X_test.drop('INR_per_USD', axis=1, inplace=True)
```

```python
import numpy as np
from sklearn import metrics, linear_model

# Create linear regression object
regr = linear_model.LinearRegression()

# Train the model using the training sets
regr.fit(X_train, y_train)

predictions=regr.predict(X_test)

print("############### Predictions #################")
#print(predictions)
print("#############################################")

print("MAE = {0}".format(metrics.mean_absolute_error(y_test,predictions)))

print("MSE = {0}".format(metrics.mean_squared_error(y_test,predictions)))
```

```python
from sklearn.neural_network import MLPRegressor
from sklearn.model_selection import train_test_split

model=MLPRegressor(random_state=2, max_iter=5000, hidden_layer_sizes=(65), activation='tanh', learning_rate='invscaling')
regr = model.fit(X_train, y_train)
predictions=regr.predict(X_test)
print("MAE = {0}".format(metrics.mean_absolute_error(y_test,predictions)))

print("MSE = {0}".format(metrics.mean_squared_error(y_test,predictions)))
```
